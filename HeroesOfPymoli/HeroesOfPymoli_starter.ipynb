```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data['SN'].unique())
players_df = pd.DataFrame({"Total Players": [total_players]})
players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items = len(purchase_data['Item ID'].unique())
avg_price = purchase_data['Price'].mean()
total_purchases = len(purchase_data['Purchase ID'])
revenue = sum(purchase_data['Price'])

total_analysis_df = pd.DataFrame({
    'Unique items': [items],
    'Average Price': ['${:,.2f}'.format(avg_price)],
    'Total Purchases': [total_purchases],
    'Total Revenue': ['${:,.2f}'.format(revenue)]
    })
total_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
unique_players = purchase_data.drop_duplicates('SN')
gender_count = unique_players['Gender'].value_counts()
gender_percentage = (unique_players['Gender'].value_counts()/total_players)*100

gender_analysis_df = pd.DataFrame({
    'Players Count': gender_count,
    'Total Players Percentage': gender_percentage.map('{:.2f}%'.format)
    })
gender_analysis_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_data = purchase_data.groupby('Gender')
gender_purchase_count = gender_purchase_data['Gender'].count()
gender_average_purchase = gender_purchase_data['Price'].mean()
gender_purchase_total = gender_purchase_data['Price'].sum()
gender_average_purchase_per_person = gender_purchase_total/gender_count

gender_purchasing_analysis_df = pd.DataFrame({
    'Purchase Count': gender_purchase_count,
    'Purchase Average Price': gender_average_purchase.map('${:,.2f}'.format),
    'Total Revenue': gender_purchase_total.map('${:,.2f}'.format),
    'Average Purchase Total Per Person': gender_average_purchase_per_person.map('${:,.2f}'.format)
    })
gender_purchasing_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

purchase_data['Age Group'] =  pd.cut(purchase_data['Age'], bins = bins, labels = labels)
players_grouped_by_age = purchase_data.groupby('Age Group').SN.nunique()

age_demographics_df = pd.DataFrame({
    'Total Count': players_grouped_by_age,
    'Percentage of Players': (players_grouped_by_age / total_players * 100).map('{:.2f}%'.format)
    })
age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
unique_players = purchase_data.drop_duplicates('SN')
age_group_count = unique_players['Age Group'].value_counts()
total_purchase_per_age_group = purchase_data.groupby('Age Group').count()
total_purchase_total_per_age_group = purchase_data.groupby('Age Group').sum()

age_purchasing_analysis_df = pd.DataFrame({
    'Purchase Count': total_purchase_per_age_group['Purchase ID'],
    'Purchase Average Price': (total_purchase_total_per_age_group['Price'] / total_purchase_per_age_group['Purchase ID']).map("${:.2f}".format),
    'Total Revenue': total_purchase_total_per_age_group['Price'].map("${:.2f}".format),
    'Average Purchase Total Per Person': (total_purchase_total_per_age_group['Price'] / age_group_count).map("${:.2f}".format),
})
age_purchasing_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_count_per_SN = purchase_data["SN"].value_counts()
average_purchase_price_per_SN = purchase_data["Price"].groupby(purchase_data["SN"]).mean()
total_purchase_value_per_SN = purchase_data["Price"].groupby(purchase_data["SN"]).sum()

top_spenders_df = pd.DataFrame({
    "Purchase Count": purchase_count_per_SN,
    "Average Purchase Price": average_purchase_price_per_SN,
    "Total Purchase Value": total_purchase_value_per_SN
    })
    


Top_5_spenders = top_spenders_df.sort_values("Total Purchase Value", ascending= False)
Top_5_spenders["Average Purchase Price"] = Top_5_spenders["Average Purchase Price"].map("${:.2f}".format)
Top_5_spenders["Total Purchase Value"] = Top_5_spenders["Total Purchase Value"].map("${:.2f}".format)


Top_5_spenders[:5]
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items = purchase_data[['Item ID', 'Item Name', 'Price']]

item_purchase_data = popular_items.groupby('Item ID')
item_names = item_purchase_data['Item Name']
purchase_count = item_purchase_data['Item Name'].count()

average_item_price = popular_items["Price"].groupby(popular_items["Item ID"]).mean().map("${:,.2f}".format)
total_purchase_value_per_item = popular_items["Price"].groupby(popular_items["Item ID"]).sum().map("${:,.2f}".format)

popular_items = pd.DataFrame({
        "Item Name": item_names,
        "Purchase Count": purchase_count,
        "Average Item Price": average_item_price,
        "Total Purchase Value": total_purchase_value_per_item
    })

top_5_popular_items = popular_items.sort_values("Purchase Count", ascending= False)

top_5_popular_items[:5]
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
top_5_items = popular_items.sort_values("Total Purchase Value", ascending= False)

top_5_items[:5]
```
